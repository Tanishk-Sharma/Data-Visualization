```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# hist of a sample normal distribution:
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
np.random.seed(100)
x1 = np.random.normal(loc=25,scale=3.0,size=1000)
ax.hist(x1, bins=30)
ax.set_xlabel('X1')
ax.set_ylabel('Bin Count')
ax.set_title('Histogram of a Single Dataset')
```

```python
# boxplot of four normal distribution:
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
np.random.seed(100)
x1 = np.random.normal(loc=25,scale=3.0,size=1000)
x2 = np.random.normal(loc=35,scale=5.0,size=1000)
x3 = np.random.normal(loc=55,scale=10.0,size=1000)
x4 = np.random.normal(loc=45,scale=3.0,size=1000)
labels = ['X1', 'X2', 'X3', 'X4']
ax.boxplot([x1, x2, x3, x4], notch=True, labels=labels, patch_artist=True)
ax.set_xlabel('Dataset')
ax.set_ylabel('Value')
ax.set_title('Box plot of Multiple Datasets')
```
