```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# A straight line plot

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
t = [5, 10, 15, 20, 25]
d = [25, 50, 75, 100, 125]
plt.plot(t, d, label='d = 5t')
ax.set_title("Time vs Distance Covered")
ax.set_xlabel("time (seconds)"); ax.set_ylabel('distance (meters)')
ax.set_xlim([0,30])
ax.set_ylim([0,130])
plt.legend()
```
